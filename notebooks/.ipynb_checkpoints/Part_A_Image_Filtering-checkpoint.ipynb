# Part A — Image Filtering
## INS3155 – Computer Vision Midterm Project

Apply and compare different traditional filters to remove noise and enhance images:
- Mean filter
- Gaussian filter
- Median filter
- Laplacian sharpening
- Bilateral filter (bonus)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim
import os
import time

# Ensure results directory exists
os.makedirs('../results/part_a', exist_ok=True)

# Display settings
plt.rcParams['figure.figsize'] = (16, 10)
plt.rcParams['figure.dpi'] = 100
plt.rcParams['font.size'] = 10
```

## 1. Utility Functions

```python
def add_gaussian_noise(image, mean=0, sigma=25):
    """Add Gaussian noise to an image."""
    noise = np.random.normal(mean, sigma, image.shape).astype(np.float64)
    noisy = np.clip(image.astype(np.float64) + noise, 0, 255).astype(np.uint8)
    return noisy


def add_salt_pepper_noise(image, amount=0.05):
    """Add salt-and-pepper noise to an image."""
    noisy = image.copy()
    # Salt (white pixels)
    num_salt = int(amount * image.size / 2)
    coords_salt = tuple(np.random.randint(0, d, num_salt) for d in image.shape[:2])
    noisy[coords_salt] = 255
    # Pepper (black pixels)
    num_pepper = int(amount * image.size / 2)
    coords_pepper = tuple(np.random.randint(0, d, num_pepper) for d in image.shape[:2])
    noisy[coords_pepper] = 0
    return noisy


def compute_psnr(original, filtered):
    """Compute Peak Signal-to-Noise Ratio between two images."""
    mse = np.mean((original.astype(np.float64) - filtered.astype(np.float64)) ** 2)
    if mse == 0:
        return float('inf')
    return 20.0 * np.log10(255.0 / np.sqrt(mse))


def compute_ssim(original, filtered):
    """Compute Structural Similarity Index."""
    if len(original.shape) == 3:
        return ssim(original, filtered, channel_axis=2)
    return ssim(original, filtered)


def show_images(images, titles, rows, cols, save_path=None, suptitle=None):
    """Display images in a grid with titles."""
    fig, axes = plt.subplots(rows, cols, figsize=(4*cols, 4*rows))
    if suptitle:
        fig.suptitle(suptitle, fontsize=16, fontweight='bold')
    axes = np.array(axes).flatten()
    for i, (img, title) in enumerate(zip(images, titles)):
        if i < len(axes):
            if len(img.shape) == 3:
                axes[i].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
            else:
                axes[i].imshow(img, cmap='gray')
            axes[i].set_title(title, fontsize=9)
            axes[i].axis('off')
    # Hide unused axes
    for j in range(len(images), len(axes)):
        axes[j].axis('off')
    plt.tight_layout()
    if save_path:
        fig.savefig(save_path, bbox_inches='tight', dpi=150)
        print(f'Saved: {save_path}')
    plt.show()


print('Utility functions loaded.')
```

## 2. Load Test Images

```python
# Load test images
img1 = cv2.imread('../data/part_a/image1.jpg')    # Option 1: Text & Sharp Edges
img2 = cv2.imread('../data/part_a/image2.jpg')    # Option 2: Texture & Details

if img1 is None or img2 is None:
    raise FileNotFoundError('Test images not found in ../data/part_a/')

print(f'Image 1 shape: {img1.shape}')
print(f'Image 2 shape: {img2.shape}')

show_images([img1, img2], ['Test Image 1 (Text/Edges)', 'Test Image 2 (Texture)'], 1, 2,
            save_path='../results/part_a/original_images.png',
            suptitle='Original Test Images')
```

## 3. Add Noise

We test with two noise types:
- **Gaussian noise** (σ=25): simulates sensor noise
- **Salt-and-pepper noise** (5%): simulates transmission errors

```python
np.random.seed(42)

# Add noise to Image 1
img1_gauss = add_gaussian_noise(img1, sigma=25)
img1_sp = add_salt_pepper_noise(img1, amount=0.05)

# Add noise to Image 2
img2_gauss = add_gaussian_noise(img2, sigma=25)
img2_sp = add_salt_pepper_noise(img2, amount=0.05)

show_images(
    [img1, img1_gauss, img1_sp, img2, img2_gauss, img2_sp],
    ['Img1 Original', 'Img1 + Gaussian', 'Img1 + Salt&Pepper',
     'Img2 Original', 'Img2 + Gaussian', 'Img2 + Salt&Pepper'],
    2, 3,
    save_path='../results/part_a/noisy_images.png',
    suptitle='Original vs Noisy Images'
)
```

## 4. Apply Filters

### Filter Descriptions:
1. **Mean Filter** (`cv2.blur`): Replaces each pixel with the average of its neighbors. Simple low-pass filter.
2. **Gaussian Filter** (`cv2.GaussianBlur`): Weighted average using Gaussian kernel. Better edge preservation than mean.
3. **Median Filter** (`cv2.medianBlur`): Replaces pixel with median of neighborhood. Best for salt-and-pepper noise.
4. **Laplacian Sharpening**: Enhances edges using 2nd-order derivative. Formula: `sharpened = original - α·Laplacian(original)`
5. **Bilateral Filter** (`cv2.bilateralFilter`): Edge-preserving smoothing. Considers both spatial and intensity distance. (Bonus)

```python
def apply_all_filters(noisy_image, kernel_sizes=[3, 5, 7]):
    """Apply all 5 filters with multiple kernel sizes.
    
    Returns dict: {filter_name: {kernel_size: filtered_image}}
    """
    results = {}
    
    for k in kernel_sizes:
        # Mean filter
        mean_key = f'Mean (k={k})'
        results[mean_key] = cv2.blur(noisy_image, (k, k))
        
        # Gaussian filter
        gauss_key = f'Gaussian (k={k})'
        results[gauss_key] = cv2.GaussianBlur(noisy_image, (k, k), 0)
        
        # Median filter
        median_key = f'Median (k={k})'
        results[median_key] = cv2.medianBlur(noisy_image, k)
        
        # Laplacian sharpening
        lap_key = f'Laplacian (k={k})'
        # First smooth slightly, then sharpen
        smoothed = cv2.GaussianBlur(noisy_image, (k, k), 0)
        laplacian = cv2.Laplacian(smoothed, cv2.CV_64F, ksize=k)
        sharpened = np.clip(smoothed.astype(np.float64) - 0.5 * laplacian, 0, 255).astype(np.uint8)
        results[lap_key] = sharpened
        
        # Bilateral filter (bonus)
        bilateral_key = f'Bilateral (d={k})'
        results[bilateral_key] = cv2.bilateralFilter(noisy_image, k, 75, 75)
    
    return results


print('Filter function defined.')
```

```python
# Apply filters to Gaussian-noisy Image 1
results_img1_gauss = apply_all_filters(img1_gauss)

# Apply filters to Salt-and-Pepper-noisy Image 1
results_img1_sp = apply_all_filters(img1_sp)

print(f'Filters applied. Total results: {len(results_img1_gauss)} per noise type')
```

## 5. Visualization: Side-by-Side Comparisons

```python
# Gaussian noise: compare all filters at k=5
k = 5
filter_names = ['Mean', 'Gaussian', 'Median', 'Laplacian', 'Bilateral']
keys_k5 = [f'Mean (k={k})', f'Gaussian (k={k})', f'Median (k={k})', 
            f'Laplacian (k={k})', f'Bilateral (d={k})']

images_gauss = [img1, img1_gauss] + [results_img1_gauss[key] for key in keys_k5]
titles_gauss = ['Original', 'Gaussian Noise'] + [f'{name} (k={k})' for name in filter_names]

show_images(images_gauss, titles_gauss, 2, 4,
            save_path='../results/part_a/filters_gaussian_noise_k5.png',
            suptitle=f'Image 1 — Gaussian Noise — All Filters (k={k})')
```

```python
# Salt-and-pepper noise: compare all filters at k=5
images_sp = [img1, img1_sp] + [results_img1_sp[key] for key in keys_k5]
titles_sp = ['Original', 'Salt & Pepper'] + [f'{name} (k={k})' for name in filter_names]

show_images(images_sp, titles_sp, 2, 4,
            save_path='../results/part_a/filters_sp_noise_k5.png',
            suptitle=f'Image 1 — Salt & Pepper Noise — All Filters (k={k})')
```

```python
# Kernel size comparison: Median filter with k=3,5,7 on both noise types
median_imgs_gauss = [results_img1_gauss[f'Median (k={k})'] for k in [3, 5, 7]]
median_imgs_sp = [results_img1_sp[f'Median (k={k})'] for k in [3, 5, 7]]

all_median = [img1_gauss] + median_imgs_gauss + [img1_sp] + median_imgs_sp
all_titles = (
    ['Gaussian Noisy'] + [f'Median k={k}' for k in [3,5,7]] +
    ['S&P Noisy'] + [f'Median k={k}' for k in [3,5,7]]
)

show_images(all_median, all_titles, 2, 4,
            save_path='../results/part_a/kernel_size_comparison.png',
            suptitle='Kernel Size Effect — Median Filter')
```

## 6. Quantitative Evaluation: PSNR & SSIM Tables

```python
def evaluate_filters(original, noisy_gauss, noisy_sp, kernel_sizes=[3, 5, 7]):
    """Compute PSNR and SSIM for all filters across noise types and kernel sizes."""
    results_gauss = apply_all_filters(noisy_gauss, kernel_sizes)
    results_sp = apply_all_filters(noisy_sp, kernel_sizes)
    
    filter_types = ['Mean', 'Gaussian', 'Median', 'Laplacian', 'Bilateral']
    
    # Build evaluation table
    rows = []
    for ftype in filter_types:
        for k in kernel_sizes:
            if ftype == 'Bilateral':
                key = f'Bilateral (d={k})'
            elif ftype == 'Laplacian':
                key = f'Laplacian (k={k})'
            else:
                key = f'{ftype} (k={k})'
            
            psnr_g = compute_psnr(original, results_gauss[key])
            ssim_g = compute_ssim(original, results_gauss[key])
            psnr_sp = compute_psnr(original, results_sp[key])
            ssim_sp = compute_ssim(original, results_sp[key])
            
            rows.append({
                'Filter': ftype,
                'Kernel': k,
                'PSNR_Gauss': psnr_g,
                'SSIM_Gauss': ssim_g,
                'PSNR_SP': psnr_sp,
                'SSIM_SP': ssim_sp
            })
    
    # Also compute noisy baseline
    psnr_noisy_g = compute_psnr(original, noisy_gauss)
    ssim_noisy_g = compute_ssim(original, noisy_gauss)
    psnr_noisy_sp = compute_psnr(original, noisy_sp)
    ssim_noisy_sp = compute_ssim(original, noisy_sp)
    
    return rows, (psnr_noisy_g, ssim_noisy_g, psnr_noisy_sp, ssim_noisy_sp)


# Evaluate on Image 1
eval_rows, noisy_baseline = evaluate_filters(img1, img1_gauss, img1_sp)

# Print results
print(f"{'='*80}")
print(f"NOISY BASELINE:")
print(f"  Gaussian Noise → PSNR: {noisy_baseline[0]:.2f} dB, SSIM: {noisy_baseline[1]:.4f}")
print(f"  Salt & Pepper  → PSNR: {noisy_baseline[2]:.2f} dB, SSIM: {noisy_baseline[3]:.4f}")
print(f"{'='*80}")
print(f"{'Filter':<12} {'k':>3} {'PSNR(G)':>9} {'SSIM(G)':>9} {'PSNR(SP)':>9} {'SSIM(SP)':>9}")
print(f"{'-'*55}")
for r in eval_rows:
    print(f"{r['Filter']:<12} {r['Kernel']:>3} {r['PSNR_Gauss']:>9.2f} {r['SSIM_Gauss']:>9.4f} {r['PSNR_SP']:>9.2f} {r['SSIM_SP']:>9.4f}")
```

## 7. PSNR Bar Charts

```python
# Create PSNR comparison bar chart
filter_types = ['Mean', 'Gaussian', 'Median', 'Laplacian', 'Bilateral']
kernel_sizes = [3, 5, 7]
colors = ['#2196F3', '#4CAF50', '#FF9800', '#9C27B0', '#F44336']

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

for ax_idx, (noise_type, psnr_key) in enumerate([('Gaussian Noise', 'PSNR_Gauss'), ('Salt & Pepper', 'PSNR_SP')]):
    ax = axes[ax_idx]
    x = np.arange(len(kernel_sizes))
    width = 0.15
    
    for i, ftype in enumerate(filter_types):
        vals = [r[psnr_key] for r in eval_rows if r['Filter'] == ftype]
        bars = ax.bar(x + i * width, vals, width, label=ftype, color=colors[i], alpha=0.85)
        # Add value labels on bars
        for bar, val in zip(bars, vals):
            ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.3,
                    f'{val:.1f}', ha='center', va='bottom', fontsize=7)
    
    ax.set_xlabel('Kernel Size')
    ax.set_ylabel('PSNR (dB)')
    ax.set_title(f'PSNR Comparison — {noise_type}')
    ax.set_xticks(x + width * 2)
    ax.set_xticklabels([f'k={k}' for k in kernel_sizes])
    ax.legend(fontsize=8)
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../results/part_a/psnr_comparison_chart.png', dpi=150, bbox_inches='tight')
print('Saved: ../results/part_a/psnr_comparison_chart.png')
plt.show()
```

```python
# SSIM comparison chart
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

for ax_idx, (noise_type, ssim_key) in enumerate([('Gaussian Noise', 'SSIM_Gauss'), ('Salt & Pepper', 'SSIM_SP')]):
    ax = axes[ax_idx]
    x = np.arange(len(kernel_sizes))
    width = 0.15
    
    for i, ftype in enumerate(filter_types):
        vals = [r[ssim_key] for r in eval_rows if r['Filter'] == ftype]
        bars = ax.bar(x + i * width, vals, width, label=ftype, color=colors[i], alpha=0.85)
        for bar, val in zip(bars, vals):
            ax.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.005,
                    f'{val:.3f}', ha='center', va='bottom', fontsize=7)
    
    ax.set_xlabel('Kernel Size')
    ax.set_ylabel('SSIM')
    ax.set_title(f'SSIM Comparison — {noise_type}')
    ax.set_xticks(x + width * 2)
    ax.set_xticklabels([f'k={k}' for k in kernel_sizes])
    ax.legend(fontsize=8)
    ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../results/part_a/ssim_comparison_chart.png', dpi=150, bbox_inches='tight')
print('Saved: ../results/part_a/ssim_comparison_chart.png')
plt.show()
```

## 8. Edge Preservation Analysis

We use the Canny edge detector to visualize how each filter affects edge preservation.

```python
# Edge detection on filtered images (k=5, Gaussian noise)
k = 5
edge_original = cv2.Canny(cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY), 50, 150)
edge_noisy = cv2.Canny(cv2.cvtColor(img1_gauss, cv2.COLOR_BGR2GRAY), 50, 150)

filter_keys = [f'Mean (k={k})', f'Gaussian (k={k})', f'Median (k={k})', 
               f'Laplacian (k={k})', f'Bilateral (d={k})']
edge_filtered = []
edge_names = []

for key, name in zip(filter_keys, filter_names):
    filtered = results_img1_gauss[key]
    gray = cv2.cvtColor(filtered, cv2.COLOR_BGR2GRAY) if len(filtered.shape) == 3 else filtered
    edges = cv2.Canny(gray, 50, 150)
    edge_filtered.append(edges)
    edge_names.append(f'{name} Edges')

all_edges = [edge_original, edge_noisy] + edge_filtered
all_edge_titles = ['Original Edges', 'Noisy Edges'] + edge_names

show_images(all_edges, all_edge_titles, 2, 4,
            save_path='../results/part_a/edge_preservation.png',
            suptitle='Edge Preservation Analysis (Canny, k=5, Gaussian Noise)')
```

## 10. Summary & Analysis

### Key Findings:

| Filter | Gaussian Noise | Salt & Pepper | Edge Preservation |
|--------|---------------|---------------|-------------------|
| **Mean** | Moderate | Poor | ❌ Destroys edges |
| **Gaussian** | Good | Moderate | ⚠️ Some blurring |
| **Median** | Good | **Best** | ✅ Preserves well |
| **Laplacian** | Sharpening only | Sharpening only | ✅ Enhances edges |
| **Bilateral** | **Best** | Good | ✅ Best preservation |

### Conclusions:
1. **Median filter** is the best choice for salt-and-pepper noise — it naturally rejects outlier pixels.
2. **Bilateral filter** provides the best balance of noise reduction and edge preservation for Gaussian noise.
3. **Mean filter** is the simplest but destroys fine details and edges.
4. **Laplacian** is not a noise-reduction filter but a sharpening operator — it complements other filters.
5. Larger kernel sizes reduce more noise but increase blurring. k=5 is a good compromise.
6. **PSNR and SSIM** confirm these qualitative observations numerically.
